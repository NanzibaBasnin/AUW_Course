Part A: Detect and Handle Missing Values [Marks: 5]
Write a Python program that:
• Loads movie_ratings.csv using pandas
• Detects and prints the number of missing (NaN) values in each column
• Fills the missing values in the Rating column with the mode of that column
• Confirms there are no more NaN values in Rating and prints the updated count
• Prints the first 10 rows of the cleaned dataset

```python
#importing necessary libraries 

import pandas as pd 
import matplotlib.pyplot as plt 
```

```python

#Load the file 
df = pd.read_csv("C:\\Users\\HP\\Desktop\\Lab Sample\\movie_ratings.csv")
```

```python
#Detects and prints the number of missing (NaN) values in each column
#reference line 30 in the netflix_data_analysis.ipynb
df.isnull().sum()
```

```python
# Fills the missing values in the Rating column with the mode of that column
#reference line 4 of the netflix_cumulative_frequency.ipynb
mode_rating = df['Rating'].mode()[0]

print(f"Mode of ratings (used to fill NaN): {mode_rating}")
```

```python
#replacing the NAN values with the mode


df['Rating'].fillna(mode_rating, inplace=True)
```

```python
#checking to see that their is no Null values 
df['Rating'].isnull()
#code is refered from line 30 in the netflix_data_analysis.ipynb file
```

```python
number_null = df['Rating'].isnull().sum()
print("The number of NaN values in Ratings column is ", number_null)
```
