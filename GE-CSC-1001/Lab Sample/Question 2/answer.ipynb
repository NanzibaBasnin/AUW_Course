```python
import matplotlib.pyplot as plt

# Data
months = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]

fiction = [120, 135, 150, 140, 160, 175, 168, 155, 145, 130, 125, 180]
non_fiction = [85, 90, 95, 88, 100, 110, 105, 98, 92, 87, 80, 115]
children = [60, 72, 80, 75, 90, 95, 88, 82, 78, 65, 70, 100]

# Plot lines
plt.plot(months, fiction, marker='o', label='Fiction')
plt.plot(months, non_fiction, marker='s', label='Non-Fiction')
plt.plot(months, children, marker='^', label='Children')

# Labels and title
plt.xlabel("Months")
plt.ylabel("Number of Books Sold")
plt.title("Monthly Book Sales ~ 2024")

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Grid and legend
plt.grid()
plt.legend()

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
